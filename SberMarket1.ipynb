```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
cutoff_date = '2020-09-02'
df['id'] = df['user_id'].astype(str) + ";" + df["cart"].astype(str)
df = df.drop(['user_id','cart'], axis=1)
df_before = df.loc[df['order_completed_at'] <= cutoff_date]
df_after = df.loc[df['order_completed_at'] > cutoff_date]
```

```python
del df_before['order_completed_at']
del df_after['order_completed_at']
df_before['target']=1
df_after['target']=0
df_before = df_before.drop_duplicates(subset='id', keep='first')
df_after = df_after.drop_duplicates(subset='id', keep='first')
```

```python
result = pd.concat([df_before, df_after], join='outer', axis=0)
result = result.sort_values(by='target', ascending=False)

result = result.drop_duplicates(subset='id', keep='first')
result = result.sort_values(by='id')
result.head(40)
```
